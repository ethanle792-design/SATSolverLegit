```python
from pyeda.inter import *

# define yur variables
x1, x2, x3, x4 = map(exprvar, ['x1', 'x2', 'x3', 'x4'])

# build expression
f = (~x1 | x2) & (x3 | ~x4) & x1

# check SAT and get all solutions
print(f.satisfy_all())
```

```python
for solution in f.satisfy_all():
    print(solution)
```
